```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits
%matplotlib inline
```

```python
df = pd.read_csv("solar_desaldata v4.0.csv", encoding = "ISO-8859-1")
```

```python
def convert(v):
    try:
        x = str(v).replace(",","")
        if " " in x:
            return float(str(x).split(" ")[0])
        else:
            float(x)
    except Exception as e:
        print(e)
        return None
```

```python
df["EPC price"] = df["EPC price"].apply(convert)
df["Capacity (m3/d)"] = df["Capacity (m3/d)"].apply(convert)
df["Capital Cost per Unit of Daily Capacity ($/m3/d)"] = df["Capital Cost per Unit of Daily Capacity ($/m3/d)"].apply(convert)
```

```python
df.to_csv("DesalData_v2.csv",encoding='utf-8',index=False)
```
